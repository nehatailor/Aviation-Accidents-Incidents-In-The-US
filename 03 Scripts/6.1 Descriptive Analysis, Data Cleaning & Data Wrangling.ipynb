### 3. Data Cleaning and Wrangling

```python
# Rename columns

df.rename(columns = {'LOCATION':'Location', 'Aircraft.damage':'Aircraft.Damage', 'Broad.phase.of.flight':'Broad.Phase.Of.Flight'}, inplace = True)
```

There are many columns in this data set that are not relevant for this analysis as there is a lot of data missing, therefore I will identify them and drop those columns.

```python
# Drop irrelevant columns not required for the analysis.

df = df.drop(columns = ['Event.Id', 'Accident.Number', 'Airport.Code', 'Registration.Number', 'FAR.Description', 'Schedule', 'Air.carrier', 'Report.Status', 'Publication.Date'])
```

```python
# Check the columns have successfully been dropped

df.dtypes
```

```python
df.shape
```

The 'Location' column contains information regarding the City and State. I feel that by splitting the column will allow me to produce deeper insight in to the data set. 

```python
df['City'] = df['Location'].str.split(',').str[0]
```

```python
df['State'] = df['Location'].str.split(',').str[1]
```

The 'Injurity.Severity' column also contains information about the count of fatalities but a separate column already exists for this information. I will also split this column.

```python
df['InjurySeverityType'] = df['Injury.Severity'].str.split(',').str[0]
```

Now that the columns have been split I will check to see if they have been done.


```python
df.dtypes
```

As we can see above the last 3 columns have been created therefore I do not require the 'Location' and 'Injury.Severity' and so they will be dropped.

```python
# Drop 'Location' and 'Injury.Severity' columns

df = df.drop(columns = ['Location', 'Injury.Severity'])
```

```python
# Check columns have been removed

df.info()
```

```python
# Check for Missing Values

df.isnull().sum()
```

As we can see from the above every column part from the 'Event.Date' column has missing values. As the missing values are on the greater side I find that they will need to be replaced to give a better insight in to the analysis. 


```python
df.describe()
```

For the numerical variable I w have decide to replace them with the mean value.

```python
df['Latitude'].fillna(df['Latitude'].mean(), inplace = True)
```

```python
df['Longitude'].fillna(df['Longitude'].mean(), inplace = True)
```

```python
df['Number.of.Engines'].fillna(df['Number.of.Engines'].mean(), inplace = True)
```

```python
df['Total.Fatal.Injuries'].fillna(df['Total.Fatal.Injuries'].mean(), inplace = True)
```

```python
df['Total.Serious.Injuries'].fillna(df['Total.Serious.Injuries'].mean(), inplace = True)
```

```python
df['Total.Minor.Injuries'].fillna(df['Total.Minor.Injuries'].mean(), inplace = True)
```

```python
df['Total.Uninjured'].fillna(df['Total.Uninjured'].mean(), inplace = True)
```

```python
# check if the above columns if the missing values have been replaced

df.isnull().sum()
```

The columns have been successfully replaced. The categorical variables still need to be replaced. For this I have decided to replace them with NaN.

```python
df.dtypes
```

```python
# Replace categorical columns with NaN

categorical_columns = df.select_dtypes(include=['object']).columns

df[categorical_columns] = df[categorical_columns].fillna("NaN")
```

```python
# Check missing values have been successfully replaced

df.isnull().sum()
```

As we can see there are no more missing values.

```python
# Check for duplicates

df_dups = df[df.duplicated()]
```

```python
df_dups
```

There are no duplicates in the data set as the event dates differ.

```python
# Check for any mixed-type data columns

for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

There are no mixed type data columns in this data set
