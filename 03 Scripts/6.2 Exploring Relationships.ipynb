### Variables suitable for Exploratory Visual Analysis

```python
# Create subsets of variables used for the analysis

sub = df[['Country', 'Latitude', 'Longitude', 'Airport.Name', 'Aircraft.Damage', 'Aircraft.Category',
       'Make', 'Model', 'Amateur.Built', 'Number.of.Engines', 'Engine.Type',
       'Purpose.of.flight', 'Total.Fatal.Injuries', 'Total.Serious.Injuries',
       'Total.Minor.Injuries', 'Total.Uninjured', 'Weather.Condition',
       'Broad.Phase.Of.Flight', 'City', 'State', 'InjurySeverityType']]
```

```python
sub
```

```python
# Create a subplot with matplotlib.

f,ax = plt.subplots(figsize=(15,15))
```

```python
# Create correlation heatmap in seaborn based on the correlation matrix and subplots

corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

```python
corr.figure
```

```python
# Export correlation heat map.

corr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'heatmap.png'))
```

For the Number of Engines and Total Uninjured, there seems to be a moderate relationship between the 2 variables. So it may be safe to assume that the number of engines relates to the size of the aircraft and therefore affects the number of injured/uninjured during the accident or incident.

There is a weak relationship between the Total Fatal Injuries and Total Serious Injuries which could indicate that out of the higher number of serious injuries, the lower the fatal injuries or the higher the fatal injuries, the lower the serious injuries. 

There is also a weak relationship between the Total Serious Injuries and Total Minor Injuries which could indicate that out of the higher number of serious injuries, the lower the minor injuries or the higher the minor injuries, the lower the serious injuries. 

Most of the other variables against each other are very to having no relationship with each other.

### Scatterplot(s):

```python
# Create a scatterplot for 'Number of Engines' and 'Total Uninjured' in Seaborn

scatterplot = sns.lmplot(x = 'Total.Uninjured', y = 'Number.of.Engines', data = df)
```

```python
# Export scatterplot

scatterplot.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'engines_uninjured.png'))
```

From this scatterplot we can see the points are structured and there seems to be a steep, positive relationship between the two variables. We can assume that as the number of the Total Uninjured persons increase the number of engines decrease, meaning the larger the aircraft the greater the number of survivors or uninjured persons.

However, there many points that are away from the line and a significant number of outliers. Becasue of the many points away the correlation coefficient alone is not sufficient to provide an accurate view of the relationship between the variables. 
It gives me an opportunity to delve deeper in to the results and adata set and see why there are many outliers its causes, does the size of the aircraft impact the number of accidents/incidents or fatalities.

### Categorical Plot:

Use histograms to visualize the distirbution of variables

```python
Hist = sns.distplot(df['Number.of.Engines'], bins = 20)
```

```python
# Export Histogram

Hist.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'num_of_engines.png'))
```

Create categories:

```python
df.loc[df['Number.of.Engines'] < 2, 'Aircraft Size'] = 'Small Aircraft'
```

```python
df.loc[(df['Number.of.Engines'] >= 2) & (df['Number.of.Engines'] < 6), 'Aircraft Size'] = 'Medium Aircraft'
```

```python
df.loc[df['Number.of.Engines'] >= 6, 'Aircraft Size'] = 'Large Aircraft'
```

```python
df['Aircraft Size'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the Aircraft sizes above

sns.set(style = 'ticks')
aviation = sns.catplot(x="Total.Fatal.Injuries", y="Number.of.Engines", hue="Aircraft Size", data=df)
```

```python
# Export categorical plot

aviation.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'categorical_plot.png'))
```

Most fatalities occur in smaller aircrafts with a single engine followed by injuries occuring on medium sized aircrafts that have engine between 2 and 4.
